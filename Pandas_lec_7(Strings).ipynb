```python
import numpy as np
import pandas as pd
```

```python
# What are vectorized operations
a=np.array([1,2,3,4])
a*4
```

```python
# problem in vectorized opertions in vanilla python
s=['cat','mat',None,'rat']
[i.startswith('c') for i in s]  # since there was a none (which is not a string)
# also it is very slow process
```

```python
# How pandas solves this issue?
s=pd.Series(['cat','mat',None,'rat'])
# string accessor
s.str.startswith('c')
#fast and optimize
```

```python
#import titanic
df=pd.read_csv('/content/titanic.csv')
df['Name']
```

```python
#common Functions
#lower/upper/capitalize/title
df['Name'].str.lower()
df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title()
#len
df['Name'].str.len()
#strip  (unnecessray spaces remove)
df['Name'].str.strip()
```

```python
df['Name'][df['Name'].str.len()==82].values[0]
```

```python
#split->get
df['lastname']=df['Name'].str.split(',').str.get(0)
df.head()
```

```python
#To use vectorized string operation always use str
```

```python
df[['title','firstname']]=df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
# n =1 means gap is 1 , expand converted series to data frame
df.head()
df['title'].value_counts()
```

```python
#replace
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
df['title'].value_counts()
```

```python
#filtering
df[df['title'].str.contains('Miss')]
#startswith/endswith
df[df['firstname'].str.startswith('A')]
#isdigit/isalpha...
df[df['lastname'].str.isdigit()]
```

```python
#applying regex (regular expression)
#contains
#search john->both case
df[df['firstname'].str.contains('john',case=False)]
#find lastname with start and end char vowel
df[df['lastname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```
