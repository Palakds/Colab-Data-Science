```python
# aggregate method
#passing dict
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'sum',
        'No_of_Votes':'max',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# Assuming 'Title' is the column with the movie titles
genres[['Runtime', 'IMDB_Rating', 'No_of_Votes', 'Gross', 'Metascore']].agg(['min','max','mean'])
```

```python
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'sum',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
#looping on groups
for group,data in genres:
  print(type(group),type(data))
  print(data)
```

```python
#find the highest rated movie of each genre using loop
df=pd.DataFrame(columns=movies.columns)
print(df)
temp_df_list = [] # create a list to hold the highest rated movies
for group,data in genres:
  temp_df_list.append(data[data['IMDB_Rating']==data['IMDB_Rating'].max()]) # append the highest rated movie to the list
df = pd.concat(temp_df_list) # concatenate all the DataFrames in the list into a single DataFrame
df
```

```python
#split(apply) combine
#apply -> builtin function
genres.apply(min)
```

```python
# find number of movies starting with A for each group
def foo(group):
  print(group)
  return group
```

```python
genres.apply(foo)
```

```python
def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
# find ranking of each movie in the group according to IMDB score
def rank_movie(group):
  group['genre_rank']=group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group wise
def normal(group):
  group['norm_rating']=(group['IMDB_Rating']-group['IMDB_Rating'].min()/group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
  return group
```

```python
genres.apply(normal)
```

```python
# groupby on multiple cols
duo=movies.groupby(['Director','Star1'])
duo
#size
duo.size()
#get_groups
duo.get_group(('Aamir Khan','Amole Gupte'))
```

```python
# find the most earning actor->director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)
```

```python
# aggregate on multiple groupby
# Select only numeric columns before aggregation
duo[[ 'Gross', 'Runtime', 'Metascore']].agg(['min','max','mean'])
```

```python
ipl=pd.read_csv('/content/deliveries.csv')
ipl.head()
```

```python
ipl.shape
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
six=ipl[ipl['batsman_runs']==6]
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
temp_df=ipl[ipl['over']>15]
temp_df=temp_df[(temp_df['batsman_runs']==4)|(temp_df['batsman_runs']==6)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find V Kohli's record against all teams
vk=ipl[ipl['batsman']=='V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return the highest score of any batsman
def highest_score(batsman):
  return ipl[ipl['batsman']==batsman]['batsman_runs'].sum()
```

```python
highest_score('V Kohli')
```
